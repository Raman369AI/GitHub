```python
# Assuming 'mushrooms' is your DataFrame
i_list = {}
for i in mushrooms.columns:
    if mushrooms[i].dtype == object :
        i_list[i] = mushrooms[i].value_counts().to_dict()
        print(i_list[i])
```

```python
filtered_i_list = {col: {value: count for value, count in counts.items() if count >= 10} for col, counts in i_list.items()}

print(filtered_i_list)
```

```python
for i in filtered_i_list:
    mushrooms[i] = mushrooms[i].apply(lambda x: x if x in filtered_i_list[i].keys() else np.nan)
```

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.offline as pyo

# Assuming mushrooms is your DataFrame
for i in mushrooms.columns:
    if mushrooms[i].dtype == object:
        layout = go.Layout(
            title=f'Poisonous/edible - {i}',
            xaxis=dict(title=f'Feature name: {i}'),
            yaxis=dict(title='Count'),
            hovermode='closest',
            autosize=True
        )

        # Get value counts for edible and poisonous classes
        edible_counts = mushrooms[mushrooms['class'] == 'e'][i].value_counts()
        poisonous_counts = mushrooms[mushrooms['class'] == 'p'][i].value_counts()

        # Align the indices of both Series
        all_values = edible_counts.index.union(poisonous_counts.index)
        edible_counts = edible_counts.reindex(all_values, fill_value=0)
        poisonous_counts = poisonous_counts.reindex(all_values, fill_value=0)

        trace0 = go.Bar(x=all_values, y=edible_counts, name='Edible', marker={'color': 'blue'})
        trace1 = go.Bar(x=all_values, y=poisonous_counts, name='Poisonous', marker={'color': 'red'})

        data = [trace0, trace1]
        fig = go.Figure(data=data, layout=layout)
        pyo.plot(fig)
```

```python
Y_train = mushrooms['class'].apply(lambda x: 1 if x=='p' else 0)
X_train = mushrooms.iloc[:,2:]
   
```

```python
X_train
```

```python
from sklearn.preprocessing import TargetEncoder
```

```python
enc_auto = TargetEncoder(smooth="auto")
```

```python
X_trans = enc_auto.fit_transform(X_train, Y_train)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_trans, Y_train, test_size=0.3, random_state=42)#%%

import lightgbm as lgb
```

```python
train_data = lgb.Dataset(X_trans, label=Y_train)
```

```python
test_data = lgb.Dataset(X_test, label=y_test, reference=train_data)
```

```python
params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'boosting_type': 'gbdt',  # Gradient Boosting Decision Tree
    'num_leaves': 31,         # Number of leaves in one tree
    'learning_rate': 0.05,
    'feature_fraction': 0.9   # Randomly select a fraction of features to build each tree
}
```

```python
lgb_model = lgb.train(
    params,
    train_data,
    num_boost_round=20,
    valid_sets=[train_data, test_data]
)
```

```python
y_pred_prob = lgb_model.predict(X_test, num_iteration=lgb_model.best_iteration)
```

```python
y_pred = [1 if x > 0.5 else 0 for x in y_pred_prob]
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_prob)

print(f"Accuracy: {accuracy}")
print(f"ROC AUC: {roc_auc}")
```

```python
cv_results = lgb.cv(
    params,
    train_data,
    nfold = 5,
    num_boost_round=100,  # Number of boosting rounds  # Number of folds
    stratified=True,  # Stratified folds (ensures even distribution of labels)
    shuffle=True,  # Shuffle data before splitting into folds
    metrics='binary_error',  # Evaluation metric  # Stop early if no improvement
    
)

# Extract best number of boosting rounds
```
