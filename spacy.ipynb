#If you remove the @Language.component decorator:
No Registration:
The function remove_gpe will no longer be registered as a pipeline component.
You cannot add it to the pipeline using nlp.add_pipe("remove_gpe"), because spaCy won't recognize "remove_gpe" as a valid component.
Manual Invocation Only:
Without registration, you can still use the function manually by passing a Doc object to it:

````python
@Language.component("remove_gpe")
def remove_gpe(doc):
  #generator to a list
  original_ents = list(doc.ents)
  for ent in doc.ents:
    if ent.label_ == "GPE":
      original_ents.remove(ent)
  doc.ents = original_ents
  return (doc)
````

In spaCy, a **Span** is a contiguous slice of tokens from a `Doc` object, representing a segment of the text. Spans are useful for working with specific portions of text, such as phrases, sentences, or named entities. Here's an overview of the **use and functionality of spans in spaCy**:

---

## **Uses of Spans in spaCy**

### **1. Representing Subsections of Text**
- A span is essentially a slice of a `Doc` object. It allows you to focus on specific parts of the text without modifying the original document.
- Example:
  ```python
  doc = nlp("I live in New York.")
  span = doc[3:5]  # Tokens 3 and 4
  print(span.text)  # Output: "New York"
  ```

### **2. Named Entity Recognition (NER)**
- Spans are commonly used to represent named entities detected in the text. Each entity is stored as a `Span` object within the `doc.ents` property.
- Example:
  ```python
  for ent in doc.ents:
      print(ent.text, ent.label_)  # Outputs entities and their labels
  ```

### **3. Creating Custom Spans**
- You can manually create spans and assign attributes like labels to them. This is useful for adding custom entities or phrases that are not recognized by the default pipeline.
- Example:
  ```python
  from spacy.tokens import Span
  span = Span(doc, start=3, end=5, label="GPE")  # "New York" as GPE
  doc.ents = list(doc.ents) + [span]  # Add custom span to entities
  ```

### **4. Sentence Segmentation**
- Spans are used to represent sentences when iterating over `doc.sents`. Each sentence is a `Span` object.
- Example:
  ```python
  for sentence in doc.sents:
      print(sentence.text)
  ```

### **5. Linguistic Features and Properties**
Spans provide access to linguistic features such as:
- **Root Token**: The syntactic root of the span.
- **Lefts and Rights**: Tokens to the left or right of the span's head.
- Example:
  ```python
  span = doc[3:5]
  print(span.root.text)   # Root token of the span
  print([t.text for t in span.lefts])   # Tokens to the left of the root
  ```

### **6. Similarity Comparisons**
- Spans can be compared for similarity with other spans, tokens, or documents using word vectors.
- Example:
  ```python
  span1 = doc[0:2]
  span2 = doc[3:5]
  print(span1.similarity(span2))  # Semantic similarity score
  ```

### **7. Custom Processing**
- Spans can be used for custom processing tasks such as extracting noun phrases, filtering tokens, or summarizing sections of text.

---

## **Why Use Spans Instead of Slicing Tokens?**
While slicing tokens (`doc[start:end]`) also produces a span-like object, using spans provides additional functionality:
1. **Attributes and Methods**: Spans have properties like `.root`, `.ents`, `.similarity()`, `.start`, `.end`, etc., which are not available with basic slices.
2. **Integration with spaCy Pipelines**: Spans integrate seamlessly with spaCy's linguistic annotations (e.g., dependency parsing, NER).
3. **Custom Labels**: You can assign custom labels to spans (e.g., marking specific phrases as entities).

---

## **Examples from Literature**

### Example: Creating and Labeling a Span [4]
```python
from spacy.tokens import Span

doc = nlp("I live in New York.")
span = Span(doc, start=3, end=5, label="GPE")  
print(span.text)         # Output: "New York"
print(span.label_)       # Output: "GPE"
```

```python
    print(sentence.text)
```
