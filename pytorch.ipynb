---

### **1. `torch.tensor()` Always Copies Data**
When you use `torch.tensor()` to create a tensor, it **always creates a new tensor by copying the data** from the input. This ensures that the resulting tensor is independent of the original data. For example:

```python
import torch
import numpy as np

# Using torch.tensor()
arr = np.array([1, 2, 3])
tensor = torch.tensor(arr)
arr[0] = 99  # Modify the original array
print(tensor)  # Output: tensor([1, 2, 3]) -> No change in tensor
```

Here, `torch.tensor()` copies the data from `arr`, so any changes to `arr` do not affect the tensor. This is useful when you want to ensure that the tensor is independent of its source.

However, copying data can be **inefficient** if you don't need a new copy and just want to work with the same memory.

---

### **2. Avoiding Copies with `torch.as_tensor()`**
If you have a NumPy array (or other array-like objects) and want to create a tensor **without copying data**, you can use `torch.as_tensor()`. This function shares memory with the input data, meaning changes to one will reflect in the other.

Example:
```python
# Using torch.as_tensor()
arr = np.array([1, 2, 3])
tensor = torch.as_tensor(arr)
arr[0] = 99  # Modify the original array
print(tensor)  # Output: tensor([99, 2, 3]) -> Tensor reflects changes in arr
```

**Key Use Case**: Use `torch.as_tensor()` when you want to avoid memory overhead and are okay with shared memory between the NumPy array and the tensor.

---

### **3. Changing `requires_grad` Without Copying**
If you already have a PyTorch tensor and just want to change its `requires_grad` flag (which determines whether gradients are tracked for autograd), creating a new tensor with `torch.tensor()` would be inefficient because it involves copying data.

Instead, you can use:
- **`requires_grad_()`**: Modifies the tensor *in-place* to enable or disable gradient tracking.
- **`detach()`**: Returns a new view of the same data without gradient tracking.

#### Example: Using `requires_grad_()`
```python
tensor = torch.randn(3, requires_grad=False)
tensor.requires_grad_()  # Enable gradient tracking in-place
print(tensor.requires_grad)  # Output: True
```

#### Example: Using `detach()`
```python
tensor = torch.randn(3, requires_grad=True)
detached_tensor = tensor.detach()  # Create a view without gradient tracking
print(detached_tensor.requires_grad)  # Output: False
```

- **`requires_grad_()`** is used when you want to modify an existing tensor's gradient behavior.
- **`detach()`** is used when you want a new view of the same underlying data but without gradient tracking.

---

### **Summary of Use Cases**

| **Function**               | **Purpose**                                                                                     | **Copy Data?** |
|----------------------------|-------------------------------------------------------------------------------------------------|----------------|
| `torch.tensor()`           | Creates a new tensor from input data (always copies).                                           | Yes            |
| `torch.as_tensor()`        | Converts input (e.g., NumPy array) into a tensor without copying (shares memory).               | No             |
| `requires_grad_()`         | Modifies an existing tensor's `requires_grad` flag in-place.                                    | No             |
| `detach()`                 | Creates a new view of a tensor without gradient tracking (shares memory with original).         | No             |

---

### **Why Avoid Copies?**
- Copying data consumes additional memory and computational resources.
- When working with large datasets or tensors, avoiding unnecessary copies can significantly improve performance.
- Using functions like `torch.as_tensor()`, `requires_grad_()`, or `detach()` helps optimize workflows by reusing existing memory instead of duplicating it.

By understanding these distinctions, you can write more efficient PyTorch code tailored to your specific needs!

In PyTorch, **scalars** and **vectors** are basic types of tensors that represent single values or one-dimensional arrays of values, respectively. Here's an explanation with examples:

---

## **1. Scalar in PyTorch**
A **scalar** is a tensor with zero dimensions (`dim=0`). It represents a single numerical value.

### **Characteristics of Scalars:**
- Shape: `torch.Size([])` (empty shape).
- Number of elements: 1.
- No axes or dimensions.

### **Creating Scalars:**
You can create a scalar using `torch.tensor()` with a single value:
```python
import torch

# Scalar tensor
scalar = torch.tensor(5)
print(scalar)            # Output: tensor(5)
print(scalar.shape)      # Output: torch.Size([])
print(scalar.dim())      # Output: 0 (zero dimensions)
```

### **Operations with Scalars:**
Scalars can be used in arithmetic operations with other tensors:
```python
# Arithmetic operation
tensor = torch.tensor([1, 2, 3])
result = tensor + scalar
print(result)            # Output: tensor([6, 7, 8])
```

---

## **2. Vector in PyTorch**
A **vector** is a one-dimensional tensor (`dim=1`). It represents an array of values arranged in a single row or column.

### **Characteristics of Vectors:**
- Shape: `(N,)`, where $$ N $$ is the number of elements.
- Number of dimensions: 1.
- Contains multiple elements but only one axis (dimension).

### **Creating Vectors:**
You can create a vector using `torch.tensor()` with a list of values:
```python
# Vector tensor
vector = torch.tensor([1, 2, 3, 4])
print(vector)            # Output: tensor([1, 2, 3, 4])
print(vector.shape)      # Output: torch.Size([4])
print(vector.dim())      # Output: 1 (one dimension)
```

### **Operations with Vectors:**
Vectors can be used in element-wise operations or dot products:
```python
# Element-wise addition
vector2 = torch.tensor([10, 20, 30, 40])
result = vector + vector2
print(result)            # Output: tensor([11, 22, 33, 44])

# Dot product
dot_product = torch.dot(vector.float(), vector2.float())
print(dot_product)       # Output: tensor(300.)
```

---

## **Comparison Table**

| **Type**   | **Description**                     | **Shape**       | **Example**                  |
|------------|-------------------------------------|-----------------|------------------------------|
| Scalar     | A single numerical value            | `torch.Size([])`| `torch.tensor(5)`            |
| Vector     | A one-dimensional array of values   | `(N,)`          | `torch.tensor([1, 2, 3])`    |

---

## **Key Differences**

1. **Dimensions**:
   - A scalar has no dimensions (`dim=0`), while a vector has one dimension (`dim=1`).

2. **Shape**:
   - A scalar's shape is empty (`torch.Size([])`), whereas a vector's shape is `(N,)`, where $$ N $$ is the number of elements.

3. **Use Cases**:
   - Scalars are used for single-value computations (e.g., loss values).
   - Vectors are used for arrays of data (e.g., feature vectors in machine learning).

---


In PyTorch, you can retrieve individual elements or perform operations on **scalars** and **vectors** using indexing, slicing, and other tensor operations. Below are examples of how to retrieve individual elements or subsets of data and perform operations on them.

---

## **Retrieving Individual Elements**

### **1. Scalars**
For a scalar tensor (0-dimensional), the value can be directly accessed using `.item()`:
```python
import torch

# Scalar tensor
scalar = torch.tensor(5)

# Retrieve the scalar value
value = scalar.item()
print(value)  # Output: 5
```

### **2. Vectors**
For a vector (1-dimensional tensor), you can retrieve individual elements using indexing:
```python
# Vector tensor
vector = torch.tensor([10, 20, 30, 40])

# Retrieve individual elements
print(vector[0])  # Output: tensor(10)
print(vector[2])  # Output: tensor(30)

# Convert to Python scalar
element = vector[2].item()
print(element)    # Output: 30
```

---

## **Retrieving Subsets of a Vector**

### **1. Slicing**
You can retrieve multiple elements (a slice) from a vector:
```python
# Vector slicing
subset = vector[1:3]  # Retrieves elements at indices 1 and 2
print(subset)         # Output: tensor([20, 30])
```

### **2. Indexing with Lists**
You can retrieve specific elements using a list of indices:
```python
indices = [0, 3]
selected_elements = vector[indices]
print(selected_elements)  # Output: tensor([10, 40])
```

---

## **Performing Operations on Scalars and Vectors**

### **1. Element-wise Operations**
You can perform operations like addition, subtraction, multiplication, or division between scalars and vectors:
```python
scalar = torch.tensor(2)
vector = torch.tensor([1, 2, 3])

# Scalar-vector operations
result_add = vector + scalar       # Add scalar to each element in the vector
result_mul = vector * scalar       # Multiply each element by the scalar

print(result_add)                  # Output: tensor([3, 4, 5])
print(result_mul)                  # Output: tensor([2, 4, 6])
```

### **2. Reduction Operations**
Reduction operations like sum, mean, or max can be applied to vectors:
```python
# Reduction operations on vectors
vector = torch.tensor([10, 20, 30])

sum_result = vector.sum()          # Sum of all elements
mean_result = vector.mean()        # Mean of all elements
max_result = vector.max()          # Maximum element

print(sum_result.item())           # Output: 60
print(mean_result.item())          # Output: 20.0
print(max_result.item())           # Output: 30
```

---

## **Advanced Operations**

### **1. Conditional Selection**
You can select elements from a vector based on a condition:
```python
vector = torch.tensor([10, -20, 30])

# Select positive values only
positive_values = vector[vector > 0]
print(positive_values)             # Output: tensor([10, 30])
```

### **2. Modifying Elements**
You can modify specific elements in a vector:
```python
vector = torch.tensor([1, 2, 3])

# Modify an element at index 1
vector[1] = 99
print(vector)                      # Output: tensor([1, 99, 3])
```

### **3. Dot Product (Vector-Vector Operation)**
The dot product is computed as the sum of element-wise products:
```python
vector1 = torch.tensor([1, 2, 3])
vector2 = torch.tensor([4, 5, 6])

dot_product = torch.dot(vector1.float(), vector2.float())
print(dot_product.item())          # Output: 32.0 (1*4 + 2*5 + 3*6)
```

### **4. Norm (Magnitude of a Vector)**
The norm of a vector is its magnitude (Euclidean distance):
```python
vector = torch.tensor([3.0, 4.0])

norm = torch.norm(vector)
print(norm.item())                 # Output: 5.0 (sqrt(3^2 + 4^2))
```

---

## **Combining Scalars and Vectors**

You can combine scalars and vectors for more complex operations:
```python
scalar = torch.tensor(5)
vector = torch.tensor([10, -15, 20])

# Add scalar to positive values only in the vector
result = torch.where(vector > 0, vector + scalar, vector)
print(result)                      # Output: tensor([15, -15, 25])
```

Here’s how you can create tensors filled with zeros, ones, and random numbers in PyTorch, along with an explanation of the code:

---

## **Code and Explanation**

```python
import torch

# 1. Creating a tensor of zeros
zeros_tensor = torch.zeros(3, 4)  # Creates a 3x4 tensor filled with zeros

ones_tensor = torch.ones(2, 5)  # Creates a 2x5 tensor filled with ones
```

### **Explanation**

   - Creates a tensor filled with zeros.
   - The shape is specified as arguments (e.g., `(3, 4)` for a 3x4 tensor).

   ```python

   ```

   - Creates a tensor filled with ones.
   - Similar to `torch.zeros()`, but all elements are set to one.
   - Often used for initializing weights or bias terms in machine learning models.
   ```python
   ones_tensor = torch.ones(2, 5)
   ```

3. **`torch.rand()`**:
   - Generates a tensor with random numbers drawn from a **uniform distribution** in the range $$[0, 1)$$.
   - Each element is independently sampled.
   - Commonly used for tasks like initializing weights randomly.
   ```python
   random_tensor_uniform = torch.rand(3, 3)
   ```

4. **`torch.randn()`**:
   - Generates a tensor with random numbers drawn from a **normal distribution** (Gaussian) with mean $$0$$ and standard deviation $$1$$.
   - Useful in scenarios where normally distributed data is required, such as initializing weights in deep learning models.

Tensor manipulation is a core aspect of PyTorch, enabling efficient handling of multi-dimensional data for deep learning and other computational tasks. Below is a **deep explanation of tensor manipulations** with examples, categorized into key operations.

---

## **1. Reshaping Tensors**

Reshaping allows you to change the dimensions of a tensor without altering its data.

### **Key Methods:**
- **`view()`**: Reshapes the tensor while preserving the underlying data.
- **`reshape()`**: Similar to `view()`, but can handle non-contiguous tensors.
- **`unsqueeze()`**: Adds a new dimension of size 1.
- **`squeeze()`**: Removes dimensions of size 1.

### **Examples:**
```python
import torch

# Create a tensor with 12 elements
tensor = torch.arange(12).reshape(3, 4)  # Reshape to 3x4
print("Original Tensor:\n", tensor)

# Reshape using view()
reshaped_tensor = tensor.view(4, 3)  # Reshape to 4x3
print("\nReshaped Tensor (view):\n", reshaped_tensor)

# Add a new dimension using unsqueeze()
unsqueezed_tensor = tensor.unsqueeze(0)  # Add dimension at index 0
print("\nUnsqueezed Tensor:\n", unsqueezed_tensor)

# Remove dimensions of size 1 using squeeze()
squeezed_tensor = unsqueezed_tensor.squeeze()
print("\nSqueezed Tensor:\n", squeezed_tensor)
```

---

## **2. Slicing and Indexing**

Slicing and indexing allow you to extract specific elements or sub-tensors.

### **Key Techniques:**
- Use `:` for slicing (e.g., `tensor[:, :2]` for the first two columns).
- Use integer indices for specific elements (e.g., `tensor[1, 2]`).
- Use advanced indexing with tensors as indices.

### **Examples:**
```python
# Create a tensor
tensor = torch.arange(12).reshape(3, 4)

# Slice rows and columns
sliced_tensor = tensor[:, 1:3]  # Extract columns 1 and 2
print("\nSliced Tensor:\n", sliced_tensor)

# Index specific rows or elements
indexed_element = tensor[1, 2]  # Element at row 1, column 2
print("\nIndexed Element:", indexed_element)

# Advanced indexing
advanced_indexing_tensor = tensor[torch.tensor([0, 2]), :]  # Select rows 0 and 2
print("\nAdvanced Indexing:\n", advanced_indexing_tensor)
```

---

## **3. Concatenation and Stacking**

Combining tensors along a specified dimension is essential for building larger datasets or models.

### **Key Methods:**
- **`torch.cat()`**: Concatenates tensors along an existing dimension.
- **`torch.stack()`**: Stacks tensors along a new dimension.

### **Examples:**
```python
# Create two tensors
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[5, 6], [7, 8]])

# Concatenate along rows (dim=0)
concat_tensor = torch.cat((tensor1, tensor2), dim=0)
print("\nConcatenated Tensor (dim=0):\n", concat_tensor)

# Stack along a new dimension (dim=0)
stacked_tensor = torch.stack((tensor1, tensor2), dim=0)
print("\nStacked Tensor (dim=0):\n", stacked_tensor)
```

---

## **4. Mathematical Operations**

PyTorch supports element-wise operations, reductions, and matrix operations.

### **Element-Wise Operations:**
```python

mean_result = tensor.float().mean()
print("Mean of Elements:", mean_result)
```

### **Matrix Multiplication:**
```python

```

## **5. Transposing and Permuting**

These operations rearrange the dimensions of a tensor.

### **Key Methods:**
- **`transpose()`**: Swaps two dimensions.
- **`permute()`**: Rearranges dimensions in any order.

### **Examples:**
```python
tensor = torch.randn(3, 4)

# Transpose rows and columns
transposed_tensor = tensor.transpose(0, 1)
print("\nTransposed Tensor:\n", transposed_tensor)

# Permute dimensions (useful for higher-dimensional tensors)
permuted_tensor = torch.randn(2, 3, 4).permute(2, 0, 1)  
print("\nPermuted Tensor Shape:", permuted_tensor.shape)
```

---

## **6. Masking and Conditional Selection**

Masking allows you to filter elements based on conditions.

### **Examples:**
```python
tensor = torch.tensor([10, -20, 30])

# Select positive values only
positive_values = tensor[tensor > 0]
print("\nPositive Values:\n", positive_values)

# Replace negative values with zero
replaced_tensor = torch.where(tensor < 0, torch.tensor(0), tensor)
print("\nReplaced Negative Values with Zero:\n", replaced_tensor)
```

---

## **7. Broadcasting**

Broadcasting enables operations between tensors of different shapes by automatically expanding their dimensions.

### **Example:**
```python
tensor1 = torch.tensor([[1], [2], [3]])   # Shape (3,1)
tensor2 = torch.tensor([10, 20])         # Shape (2,)
result = tensor1 + tensor2               # Broadcasts to shape (3,2)
print("\nBroadcasting Result:\n", result)
```

## **1. `torch.nn`**
The `torch.nn` module provides tools to build and train neural networks. It includes predefined layers, activation functions, loss functions, and utilities for creating models.

### **Key Features:**
- **`nn.Module`**: Base class for all neural network models.
- **Layers**: Predefined layers like `nn.Linear`, `nn.Conv2d`, etc.
- **Activation Functions**: Functions like `nn.ReLU`, `nn.Sigmoid`, etc.
- **Loss Functions**: Functions like `nn.CrossEntropyLoss`, `nn.MSELoss`.

### **Example: Building a Simple Neural Network**
```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# Define a simple neural network
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(28 * 28, 128)  # Input layer
        self.fc2 = nn.Linear(128, 64)      # Hidden layer
        self.fc3 = nn.Linear(64, 10)       # Output layer

    def forward(self, x):
        x = x.view(-1, 28 * 28)            # Flatten input
        x = F.relu(self.fc1(x))            # Apply ReLU activation
        x = F.relu(self.fc2(x))
        x = self.fc3(x)                    # No activation for output (logits)
        return x

# Instantiate the model
model = SimpleNN()
print(model)
```

## **2. `torch.optim`**
The `torch.optim` module provides optimization algorithms to update model parameters during training. These algorithms minimize the loss function by adjusting the weights and biases.

### **Key Features:**
- Built-in optimizers like SGD, Adam, RMSprop.
- Learning rate schedulers for dynamic learning rate adjustment.

### **Example: Using an Optimizer**
```python
import torch.optim as optim

# Define the model, loss function, and optimizer
model = SimpleNN()
criterion = nn.CrossEntropyLoss()  # Loss function
optimizer = optim.Adam(model.parameters(), lr=0.001)  # Adam optimizer

# Training loop
for epoch in range(5):  # Number of epochs
    inputs = torch.randn(32, 28 * 28)  # Random input (batch size: 32)
    labels = torch.randint(0, 10, (32,))  # Random labels (10 classes)

    optimizer.zero_grad()                # Reset gradients
    outputs = model(inputs)              # Forward pass
    loss = criterion(outputs, labels)   # Compute loss
    loss.backward()                     # Backward pass (compute gradients)
    optimizer.step()                    # Update parameters

    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```

```python
    def __init__(self):
        self.data = torch.randn(1000, 28 * 28)   # Random data (1000 samples)
        self.labels = torch.randint(0, 10, (1000,))  # Random labels (10 classes)
```

#Differentiation

Let’s break down the code step by step and explain what’s happening:

---

### **Code Analysis**

```python
import torch

x = torch.tensor(2.0, requires_grad=True)  # Create a tensor with gradient tracking enabled
y = x ** 2                                # Define y as a function of x (y = x^2)
y.backward()                              # Compute the gradient of y with respect to x
print(x.grad)                             # Print the gradient (dy/dx)
```

---

### **Step-by-Step Explanation**

#### **1. `torch.tensor(2.0, requires_grad=True)`**
- A scalar tensor `x` is created with a value of $$ 2.0 $$.
- The `requires_grad=True` argument tells PyTorch to track operations on this tensor so that gradients can be computed during backpropagation.

#### **2. `y = x ** 2`**
- Here, `y` is defined as $$ y = x^2 $$, which is a simple mathematical operation.
- Since `x` has `requires_grad=True`, PyTorch builds a computational graph to track this operation. The computational graph records that $$ y $$ depends on $$ x $$ through the operation $$ x^2 $$.

#### **3. `y.backward()`**
- The `.backward()` method computes the gradient of `y` with respect to all tensors that have `requires_grad=True`.
- In this case, $$ y = x^2 $$, so the derivative (gradient) of $$ y $$ with respect to $$ x $$ is:
  $$
  \frac{dy}{dx} = 2x
  $$
- PyTorch automatically computes this gradient using the chain rule and stores it in the `.grad` attribute of `x`.

#### **4. `print(x.grad)`**
- After calling `y.backward()`, the gradient of $$ y $$ with respect to $$ x $$, which is $$ 2x $$, is stored in `x.grad`.
- Substituting $$ x = 2.0 $$:
  $$
  \frac{dy}{dx} = 2(2.0) = 4.0
  $$
- Therefore, `print(x.grad)` outputs:
  ```
  tensor(4.)
  ```

---

### **Key Concepts**

1. **Computational Graph**:
   - PyTorch builds a dynamic computational graph when you perform operations on tensors with `requires_grad=True`. This graph tracks dependencies between tensors and allows automatic differentiation.

2. **Gradient**:
   - The gradient represents the rate of change of one variable with respect to another. In this case, it’s the derivative of $$ y $$ with respect to $$ x $$.

3. **Backward Propagation**:
   - The `.backward()` function computes gradients by traversing the computational graph in reverse, applying the chain rule.

4. **Accessing Gradients**:
   - Gradients are stored in the `.grad` attribute of tensors with `requires_grad=True`.

---

### **Modified Example: Multiple Operations**
Let’s extend this example to include more operations:

```python
x = torch.tensor(3.0, requires_grad=True)
z = x ** 3 + 5 * x
z.backward()
print(x.grad)
```

#### Explanation:
1. The function here is:
   $$
   z = x^3 + 5x
   $$
2. The derivative of $$ z $$ with respect to $$ x $$ is:
   $$
   \frac{dz}{dx} = 3x^2 + 5
   $$
3. Substituting $$ x = 3.0 $$:
   $$
   \frac{dz}{dx} = 3(3)^2 + 5 = 27 + 5 = 32
   $$
4. Output:
   ```
   tensor(32.)
   ```

---

### **Important Notes**

1. **Calling `.backward()` Multiple Times**:
   - By default, PyTorch clears the computational graph after calling `.backward()`. To retain the graph for further computations, use `retain_graph=True`:
     ```python
     y.backward(retain_graph=True)
     ```

2. **Gradients Accumulate**:
   - Gradients accumulate in `.grad`. To avoid this, reset gradients using:
     ```python
     x.grad.zero_()
     ```

3. **No Gradient Tracking**:
   - If you don’t need gradients (e.g., during inference), you can disable tracking using:
     ```python
     with torch.no_grad():
         z = x ** 2
     ```

## **1. Basics of Derivatives in PyTorch**

### **Key Concepts**
- **`requires_grad=True`**: When creating a tensor, setting `requires_grad=True` tells PyTorch to track operations on this tensor so that gradients can be computed.
- **Computational Graph**: PyTorch builds a graph of operations applied to tensors with `requires_grad=True`. This graph is used to compute gradients.
- **Gradient Storage**: Gradients are stored in the `.grad` attribute of the tensor.
- **Backward Propagation**: The `.backward()` method computes the gradients by traversing the computational graph in reverse.

---

### **Example: Scalar Derivative**
```python
import torch

# Define a tensor with gradient tracking
x = torch.tensor(2.0, requires_grad=True)

# Define a function y = x^2
y = x ** 2

# Compute the derivative dy/dx
y.backward()

# Print the gradient
print(x.grad)  # Output: tensor(4.)
```

#### Explanation:
1. The function is $$ y = x^2 $$.
2. The derivative is $$ \frac{dy}{dx} = 2x $$.
3. Substituting $$ x = 2 $$, we get $$ \frac{dy}{dx} = 4 $$.

---

## **2. Vector Derivatives (Jacobian)**

For vector-valued functions, PyTorch computes the **Jacobian matrix**, which contains partial derivatives of each output with respect to each input.

### Example:
```python
# Define a vector input
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)

# Define a vector function y = x^2
y = x ** 2

# Compute the gradient (Jacobian)
y.backward(torch.tensor([1.0, 1.0, 1.0]))  # Provide gradient weights for vector outputs

# Print the gradient
print(x.grad)  # Output: tensor([2., 4., 6.])
```

#### Explanation:
1. The function is $$ y_i = x_i^2 $$ for $$ i \in \{1, 2, 3\} $$.
2. The derivative is $$ \frac{\partial y_i}{\partial x_i} = 2x_i $$.
3. Gradients are computed element-wise as $$ [2(1), 2(2), 2(3)] = [2, 4, 6] $$.

---

## **3. Higher-Order Derivatives**

PyTorch supports higher-order derivatives by enabling `create_graph=True` in `.backward()` or `torch.autograd.grad()`.

### Example:
```python
# Define a tensor
x = torch.tensor(3.0, requires_grad=True)

# Define a function y = x^3
y = x ** 3

# Compute first derivative (dy/dx)
y.backward(create_graph=True)
print(x.grad)  # Output: tensor(27.)

# Compute second derivative (d^2y/dx^2)
second_derivative = torch.autograd.grad(x.grad, x)
print(second_derivative)  # Output: (tensor(18.),)
```

#### Explanation:
1. The function is $$ y = x^3 $$.
   - First derivative: $$ \frac{dy}{dx} = 3x^2 $$, which evaluates to $$ 27 $$ at $$ x=3 $$.
   - Second derivative: $$ \frac{d^2y}{dx^2} = 6x $$, which evaluates to $$ 18 $$ at $$ x=3 $$.

---

## **4. Gradients for Non-Scalar Outputs**

If the output is not scalar (e.g., a vector), you need to provide a gradient argument to `.backward()` representing the gradient of the output with respect to some scalar quantity.

### Example:
```python
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
y = x * torch.tensor([4.0, 5.0, 6.0])  # Element-wise multiplication

# Provide gradient weights for non-scalar output
y.backward(torch.tensor([1.0, 1.0, 1.0]))

print(x.grad)  # Output: tensor([4., 5., 6.])
```

#### Explanation:
The gradient weights provided to `.backward()` tell PyTorch how to combine the partial derivatives of each element in `y`.

---

## **5. Zeroing Gradients**

Gradients accumulate by default in PyTorch (i.e., they are added to existing values in `.grad`). To avoid this behavior during multiple backward passes, you must zero out gradients manually.

### Example:
```python
x = torch.tensor(3.0, requires_grad=True)

for i in range(3):
    y = x ** i
    y.backward()
    print(f"Iteration {i+1}, Gradient: {x.grad}")
    x.grad.zero_()  # Reset gradients after each iteration
```

---

## **6. Disabling Gradient Tracking**

During inference or when gradients are not needed (e.g., for saving memory), you can disable gradient tracking using `torch.no_grad()` or `detach()`.

### Example:
```python
x = torch.tensor(5.0, requires_grad=True)

# Disable gradient tracking using no_grad()
with torch.no_grad():
    y = x ** 2
    print(y.requires_grad)  # Output: False

# Detach tensor from computation graph
z = x.detach()
print(z.requires_grad)      # Output: False
```

---

## **7. Computing Gradients for Multiple Inputs**

You can compute gradients for multiple inputs by passing them to `torch.autograd.grad()`.

### Example:
```python
x = torch.tensor(1.0, requires_grad=True)
w = torch.tensor(2.0, requires_grad=True)

# Define a function z = wx^2
z = w * x ** 2

# Compute gradients dz/dw and dz/dx
gradients = torch.autograd.grad(z, [w, x])
print("dz/dw:", gradients[0])   # Output: dz/dw: tensor(1.)
print("dz/dx:", gradients[1])   # Output: dz/dx: tensor(4.)
```

#### Explanation:
- $$ z = wx^2 $$
- Partial derivatives:
   - $$ \frac{\partial z}{\partial w} = x^2 $$
   - $$ \frac{\partial z}{\partial x} = 2wx $$

---

## **8. Jacobian Matrix**

You can compute the Jacobian matrix (all partial derivatives) using `torch.autograd.functional.jacobian`.

### Example:
```python
from torch.autograd.functional import jacobian

def func(x):
    return torch.tensor([x[0]**2 + x[1], x[1]**3])

inputs = torch.tensor([1.0, 2.0], requires_grad=True)

jacobian_matrix = jacobian(func, inputs)
print("Jacobian Matrix:\n", jacobian_matrix)
```

#### Explanation:
For a vector-valued function $$ f(x) $$, the Jacobian matrix contains all partial derivatives:
$$
J_{ij} = \frac{\partial f_i}{\partial x_j}
$$

### **Higher-Order Derivatives in PyTorch**

Higher-order derivatives refer to **derivatives of derivatives**. For example, the **first derivative** of a function measures the rate of change, while the **second derivative** measures the rate of change of the first derivative (e.g., curvature). PyTorch supports higher-order derivatives using its `autograd` module, which allows you to compute gradients of gradients.

---

### **Key Concepts**

1. **First Derivative**:
   - The gradient of a scalar function $$ y $$ with respect to $$ x $$, denoted as $$ \frac{dy}{dx} $$.
   - Example: If $$ y = x^2 $$, then $$ \frac{dy}{dx} = 2x $$.

2. **Second Derivative**:
   - The derivative of the first derivative, denoted as $$ \frac{d^2y}{dx^2} $$.
   - Example: If $$ y = x^2 $$, then $$ \frac{d^2y}{dx^2} = 2 $$.

3. **Higher-Order Derivatives**:
   - You can compute derivatives beyond the second order (e.g., third, fourth, etc.).
   - Example: If $$ y = x^4 $$, then:
     - First derivative: $$ \frac{dy}{dx} = 4x^3 $$
     - Second derivative: $$ \frac{d^2y}{dx^2} = 12x^2 $$
     - Third derivative: $$ \frac{d^3y}{dx^3} = 24x $$

---

### **How PyTorch Computes Higher-Order Derivatives**

PyTorch computes higher-order derivatives by enabling gradient tracking during the computation of the first derivative. This is achieved using `create_graph=True` in `.backward()` or `torch.autograd.grad()`.

---

### **Examples**

#### **1. Second Derivative Using `.backward()`**
```python
import torch

# Define a tensor with gradient tracking
x = torch.tensor(3.0, requires_grad=True)

# Define a function y = x^3
y = x ** 3

# Compute first derivative (dy/dx)
y.backward(create_graph=True)  # Retain computational graph for higher-order derivatives
print("First Derivative (dy/dx):", x.grad)  # Output: tensor(27.)

# Compute second derivative (d^2y/dx^2)
second_derivative = torch.autograd.grad(x.grad, x)
print("Second Derivative (d^2y/dx^2):", second_derivative[0])  # Output: tensor(18.)
```

#### Explanation:
1. First derivative:
   $$
   y = x^3 \implies \frac{dy}{dx} = 3x^2
   $$
   At $$ x = 3 $$, $$ \frac{dy}{dx} = 27 $$.

2. Second derivative:
   $$
   \frac{d^2y}{dx^2} = 6x
   $$
   At $$ x = 3 $$, $$ \frac{d^2y}{dx^2} = 18 $$.

---

#### **2. Third Derivative Using `torch.autograd.grad()`**
```python
# Define a tensor
x = torch.tensor(2.0, requires_grad=True)

# Define a function y = x^4
y = x ** 4

# Compute first derivative (dy/dx)
grad1 = torch.autograd.grad(y, x, create_graph=True)[0]  # Retain graph for higher-order derivatives
print("First Derivative (dy/dx):", grad1)  # Output: tensor(32.)

# Compute second derivative (d^2y/dx^2)
grad2 = torch.autograd.grad(grad1, x, create_graph=True)[0]
print("Second Derivative (d^2y/dx^2):", grad2)  # Output: tensor(48.)

# Compute third derivative (d^3y/dx^3)
grad3 = torch.autograd.grad(grad2, x)[0]
print("Third Derivative (d^3y/dx^3):", grad3)  # Output: tensor(24.)
```

#### Explanation:
1. First derivative:
   $$
   y = x^4 \implies \frac{dy}{dx} = 4x^3
   $$
   At $$ x=2 $$, $$ \frac{dy}{dx} = 32 $$.

2. Second derivative:
   $$
   \frac{d^2y}{dx^2} = 12x^2
   $$
   At $$ x=2 $$, $$ \frac{d^2y}{dx^2} = 48 $$.

3. Third derivative:
   $$
   \frac{d^3y}{dx^3} = 24x
   $$
   At $$ x=2 $$, $$ \frac{d^3y}{dx^3} = 24 $$.

---

#### **3. Jacobian and Hessian Matrix**

For vector-valued functions, higher-order derivatives can be represented using the Jacobian and Hessian matrices.

- **Jacobian**: First-order partial derivatives.
- **Hessian**: Second-order partial derivatives.

##### Example: Compute Hessian Matrix
```python
from torch.autograd.functional import hessian

# Define a function f(x)
def func(x):
    return x[0]**2 + 3 * x[1]**4

# Input tensor
inputs = torch.tensor([1.0, 1.0], requires_grad=True)

# Compute Hessian matrix
hessian_matrix = hessian(func, inputs)
print("Hessian Matrix:\n", hessian_matrix)
```

#### Explanation:
For a function $$ f(x_0, x_1) = x_0^2 + 3x_1^4 $$:
- The Hessian matrix contains all second-order partial derivatives:
$$
H =
\begin{bmatrix}
\frac{\partial^2 f}{\partial x_0^2} & \frac{\partial^2 f}{\partial x_0\partial x_1} \\
\frac{\partial^2 f}{\partial x_1\partial x_0} & \frac{\partial^2 f}{\partial x_1^2}
\end{bmatrix}
$$

---

#### **4. Higher-Order Gradients in Neural Networks**

Higher-order gradients are useful in advanced optimization techniques like meta-learning or second-order optimization.

##### Example: Computing Gradients in Neural Networks
```python
import torch.nn as nn

# Define a simple model
model = nn.Linear(1, 1)  # Single-layer linear model

# Input and target tensors
inputs = torch.tensor([[1.0]], requires_grad=True)
targets = torch.tensor([[10.0]])

# Forward pass and loss computation
outputs = model(inputs)
loss = (outputs - targets).pow(2).sum()  # Mean squared error loss

# First gradient (gradient of loss w.r.t model parameters)
first_gradients = torch.autograd.grad(loss, model.parameters(), create_graph=True)

# Second gradient (gradient of first gradients w.r.t inputs)
second_gradients = [torch.autograd.grad(g, inputs)[0] for g in first_gradients]
print("Second Gradients:", second_gradients)
```

---

### **Important Notes**
1. **`create_graph=True`**:
   - Essential for computing higher-order derivatives.
   - Retains the computational graph after computing the first gradient.

2. **Efficiency**:
   - Higher-order derivatives can be computationally expensive due to larger computational graphs.

3. **Applications**:
   - Higher-order derivatives are used in optimization techniques like Newton's method and meta-learning algorithms (e.g., MAML).

### **Using Data Transformations Inside PyTorch Custom Dataset Class**

Data transformations in PyTorch are essential for preprocessing and augmenting data during training and inference. They are typically applied within the `torch.utils.data.Dataset` class using the `transform` parameter. Below is a detailed explanation of how to use data transformations in various scenarios.

---

### **1. Basic Transformations Using `torchvision.transforms`**

The `torchvision.transforms` module provides predefined transformations for common image processing tasks like resizing, cropping, normalization, and data augmentation.

#### **Example: Applying Transformations in a Custom Dataset**
```python
import torch
from torchvision import transforms
from torch.utils.data import Dataset

# Define a custom dataset class
class CustomDataset(Dataset):
    def __init__(self, data, labels, transform=None):
        self.data = data
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = self.data[idx]
        label = self.labels[idx]
        
        # Apply the transformation if provided
        if self.transform:
            sample = self.transform(sample)
        
        return sample, label

# Example data (100 grayscale images of size 28x28)
data = torch.randn(100, 28, 28)
labels = torch.randint(0, 10, (100,))

# Define transformations
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Create the dataset with transformations
dataset = CustomDataset(data, labels, transform=transform)

# Access transformed data
sample, label = dataset[0]
print(f"Sample shape: {sample.shape}, Label: {label}")
```

#### **Explanation**:
1. **Transformations**:
   - `RandomHorizontalFlip`: Randomly flips the image horizontally.
   - `RandomRotation`: Rotates the image by a random angle.
   - `ToTensor`: Converts the image to a PyTorch tensor.
   - `Normalize`: Normalizes the image with mean and standard deviation.

2. **Transform Application**:
   - The transformations are applied in the `__getitem__()` method when accessing each sample.

---

### **2. Custom Transformations**

You can create custom transformations by defining a class with a `__call__()` method.

#### **Example: Custom Transformation**
```python
# Define a custom transformation
class MultiplyByTwo:
    def __call__(self, sample):
        return sample * 2

# Apply custom transformation
custom_transform = MultiplyByTwo()

# Create dataset with custom transformation
custom_dataset = CustomDataset(data, labels, transform=custom_transform)

# Access transformed data
sample, label = custom_dataset[0]
print(f"Transformed Sample: {sample}")
```

#### **Explanation**:
- The `MultiplyByTwo` class multiplies all pixel values in the sample by 2.
- This transformation is applied dynamically when accessing each sample.

---

### **3. Combining Predefined and Custom Transformations**

You can combine multiple transformations (both predefined and custom) using `transforms.Compose`.

#### **Example: Combined Transformations**
```python
# Combine predefined and custom transformations
combined_transform = transforms.Compose([
    MultiplyByTwo(),
    transforms.RandomHorizontalFlip(),
    transforms.Normalize((0.5,), (0.5,))
])

# Create dataset with combined transformations

# Access transformed data

print(f"Combined Transformed Sample: {sample}")
```

### **6. Advanced Use Cases**

#### **6.1 Multiple Transforms for Different Data Types**
If your dataset contains multiple types of data (e.g., images and masks), you can apply different transformations to each type.

```python
class MultiTransformDataset(Dataset):
    def __init__(self, images, masks, image_transform=None, mask_transform=None):
        self.images = images
        self.masks = masks
        self.image_transform = image_transform
        self.mask_transform = mask_transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        mask = self.masks[idx]

        if self.image_transform:
            image = self.image_transform(image)
        if self.mask_transform:
            mask = self.mask_transform(mask)

        return image, mask

# Example usage
image_transform = transforms.Compose([transforms.RandomHorizontalFlip(), transforms.ToTensor()])
mask_transform = transforms.Compose([transforms.ToTensor()])

dataset = MultiTransformDataset(data, data.clone(), image_transform=image_transform, mask_transform=mask_transform)
```

### **Why This Happens**
The `Normalize` transformation is designed for image tensors with a channel dimension. Grayscale images should have a shape of `(1, H, W)` (1 channel), but your data is missing this channel dimension and has a shape of `(H, W)` instead.

---

### **Fixing the Issue**

#### **Solution 1: Add a Channel Dimension**
You can add a channel dimension to your data dynamically using `unsqueeze()` in the `__getitem__()` method of your custom dataset.

```python
# Define a custom dataset class
class CustomDataset(Dataset):
    def __init__(self, data, labels, transform=None):
        self.data = data
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = self.data[idx].unsqueeze(0)  # Add channel dimension (C, H, W)
        label = self.labels[idx]

        if self.transform:
            sample = self.transform(sample)

        return sample, label

# Example data (100 grayscale images of size 28x28)
data = torch.randn(100, 28, 28)
labels = torch.randint(0, 10, (100,))

# Define transformations
transform = transforms.Compose([
    transforms.Normalize((0.5,), (0.5,))
])

# Create the dataset with transformations
dataset = CustomDataset(data, labels, transform=transform)

# Access transformed data
sample, label = dataset[0]
```

````python
import torch
````

```python
class ImageDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        """

            image_paths (list): List of file paths to images.

        """
        Returns the total number of samples in the dataset.
        """

        Retrieves the image and its label at the specified index.
        Args:
            idx (int): Index of the sample to retrieve.
        
        Returns:
            tuple: (image, label) where image is the transformed image tensor and label is its corresponding label.

        # Apply transformations if provided
        if self.transform:
            image = self.transform(image)
        
        # Get label
        label = self.labels[idx]
        
        return image, label
```

Yes, the parameters of a `torch.nn.Linear` layer are **initialized randomly by default**. When you create a `Linear` layer in PyTorch, the weights and biases are initialized using a uniform distribution based on the layer's input and output dimensions.

### **Default Initialization in `torch.nn.Linear`**
PyTorch initializes the weights and biases of a `Linear` layer as follows:

1. **Weights (`weight`)**:
   - The weights are initialized using a **uniform distribution**:
     $$
     \text{weight} \sim U\left(-\sqrt{\frac{1}{\text{in\_features}}}, \sqrt{\frac{1}{\text{in\_features}}}\right)
     $$
   - This range is determined by the number of input features (`in_features`) to ensure proper scaling.

2. **Biases (`bias`)**:
   - If `bias=True`, the biases are also initialized using the same uniform distribution:
     $$
     \text{bias} \sim U\left(-\sqrt{\frac{1}{\text{in\_features}}}, \sqrt{\frac{1}{\text{in\_features}}}\right)
     $$

  - $$ \eta $$ (or `lr`): Learning rate, which controls how much we adjust the weights during each step.
  - $$ \frac{\partial L}{\partial w} $$ (or `w.grad`): Gradient of the loss with respect to `w`.

#### **8. Zeroing Gradients**

This line is crucial in PyTorch training loops. Here's why:

1. **Why Zero Gradients?**
   - Gradients accumulate by default in PyTorch.
   - During backpropagation (`loss.backward()`), PyTorch adds new gradients to any existing gradients stored in `w.grad`.
   - If you don't clear (`zero_()`) these gradients after each parameter update, they will accumulate over multiple iterations, leading to incorrect updates.

   Example:
   ```python
   # Without zeroing gradients:
   grad1 = torch.tensor([0.5])
   grad2 = torch.tensor([0.3])
   accumulated_grad = grad1 + grad2  # Accumulated gradient: [0.8]
   ```

2. **How Does `.zero_()` Work?**
   - `.zero_()` is an **in-place operation** that sets all elements of a tensor to zero.
   - In this case, it clears out all accumulated gradients in `w.grad`.

3. **Why Use `.data.zero_()`?**
   - `.data` accesses the raw tensor data directly, bypassing PyTorch's autograd system.
   - This ensures that clearing gradients does not interfere with gradient tracking or computational graphs.

### **Detailed Explanation of the DataLoader Object in PyTorch**

The `torch.utils.data.DataLoader` is a core utility in PyTorch that simplifies and optimizes the process of loading data for training machine learning models. It abstracts away the complexity of batching, shuffling, and parallel data loading, making it an essential tool for handling datasets efficiently.

---

### **What is a DataLoader?**
- The `DataLoader` is an **iterable** that provides batches of data from a dataset.
- It works with a `Dataset` object (which defines how to access individual samples) and adds functionality like:
  - Automatic batching.
  - Shuffling the data.
  - Loading data in parallel using multiple worker processes.

---

### **Key Parameters of `DataLoader`**
Here are the most commonly used arguments for the `DataLoader` constructor:

1. **`dataset`**:
   - The dataset to load data from. This can be a built-in dataset (e.g., MNIST) or a custom dataset that inherits from `torch.utils.data.Dataset`.

2. **`batch_size`**:
   - Number of samples per batch.
   - Default: `1`.

3. **`shuffle`**:
   - Whether to shuffle the dataset at the start of each epoch.
   - Default: `False`.

4. **`num_workers`**:
   - Number of subprocesses to use for data loading.
   - Default: `0` (data loading happens in the main process).
   - Increasing this can speed up data loading for large datasets.

5. **`collate_fn`**:
   - A function to merge a list of samples into a batch.
   - Default: Combines samples into batches automatically.

6. **`pin_memory`**:
   - If `True`, the DataLoader will copy tensors into pinned memory (on CPU) for faster transfer to GPU.

7. **`drop_last`**:
   - If `True`, drops the last incomplete batch if the dataset size is not divisible by the batch size.

8. **Other Parameters**:
   - `sampler`, `batch_sampler`: For custom sampling strategies.
   - `timeout`: Timeout for collecting a batch from workers.

---

### **How Does DataLoader Work?**
1. The DataLoader retrieves individual samples from the dataset using its `__getitem__()` method.
2. These samples are grouped into batches based on the specified `batch_size`.
3. If shuffling is enabled, it randomizes the order of samples at each epoch.
4. If multiprocessing (`num_workers > 0`) is enabled, multiple worker processes load batches in parallel.

### **Why is `X` Being Used?**

The variable `X` represents the entire input dataset (features) and is used in two places:

1. **Before Batch Processing:**
   ```python
   Yhat = forward(X)
   ```
   - Here, predictions are computed for all samples in `X`.
   - This is done to evaluate how well the model fits all data points at this stage of training (e.g., to visualize loss across all samples).

2. **During Batch Processing:**
   ```python
   for x, y in trainloader:
       yhat = forward(x)
       ...
   ```
   - In this loop, only a subset (`x`, a batch) of the dataset is processed at a time.
   - This is part of Stochastic Gradient Descent (SGD), where updates are made based on small batches rather than the entire dataset.

---

### Summary Table

| Step                        | Purpose                                                                                     |
|-----------------------------|---------------------------------------------------------------------------------------------|
| Forward Pass                | Compute predictions (`Yhat`) based on current parameters (`w`, `b`).                        |
| Loss Calculation            | Measure how far predictions are from true labels using Mean Squared Error (MSE).           |
| Visualization               | Track parameter updates and visualize progress during training (`get_surface`).            |
| Batch Processing            | Use small subsets of data to compute gradients and update parameters iteratively (SGD).    |
| Parameter Updates           | Update weights and biases using SGD formula with learning rate (`lr`).                     |
| Clearing Gradients          | Reset accumulated gradients after each update to avoid incorrect updates in subsequent steps.|

---

### Final Notes

The use of `X` ensures that you can compute both global metrics (e.g., full dataset loss) and local metrics (e.g., batch-specific loss). This combination allows you to monitor overall progress while leveraging efficient mini-batch updates through Stochastic Gradient Descent (SGD).

The `nn.Sequential` module in PyTorch is a convenient way to define neural networks where the layers are applied sequentially. It simplifies the process of building models, especially for straightforward architectures, by eliminating the need to explicitly define a `forward()` method. Below is a detailed explanation of how to use `nn.Sequential`, along with an example.

---

### **What is `nn.Sequential`?**

`nn.Sequential` is a container module in PyTorch that stacks layers in a specified order. When input data is passed through the model, it flows sequentially through each layer in the same order as they were defined. This makes `nn.Sequential` particularly useful for linear, feed-forward architectures.

#### Advantages:
1. **Simplicity**: Reduces boilerplate code by avoiding the need to define a `forward()` method.
2. **Readability**: The sequential flow of layers makes it easy to understand the architecture.
3. **Convenience**: Ideal for prototyping simple models quickly.

#### Limitations:
- It cannot handle complex architectures that require branching or multiple inputs/outputs.
- For more advanced models, you need to subclass `nn.Module`.

---

### **How to Define a Neural Network with `nn.Sequential`**

You can define a model using `nn.Sequential` in two main ways:
1. By passing layers as arguments.
2. By using an `OrderedDict` to name the layers.

#### Example 1: Passing Layers as Arguments
```python
import torch
import torch.nn as nn

# Define a simple feedforward neural network
model = nn.Sequential(
    nn.Linear(784, 256),  # Input layer (784 features) -> Hidden layer (256 units)
    nn.ReLU(),           # Activation function
    nn.Linear(256, 10),  # Hidden layer -> Output layer (10 classes)
    nn.Softmax(dim=1)    # Softmax for probabilities
)

# Print the model architecture
print(model)
```

#### Example 2: Using an OrderedDict
```python
from collections import OrderedDict

# Define the same network with named layers
model = nn.Sequential(OrderedDict([
    ('fc1', nn.Linear(784, 256)),  # Fully connected layer 1
    ('relu1', nn.ReLU()),          # Activation function
    ('fc2', nn.Linear(256, 10)),   # Fully connected layer 2
    ('softmax', nn.Softmax(dim=1)) # Softmax for probabilities
]))

# Print the model architecture
print(model)
```

---

### **Explanation of Key Components**

1. **Layers**:
   - `nn.Linear`: Defines fully connected layers with specified input and output sizes.
   - `nn.ReLU`: Applies the ReLU activation function.
   - `nn.Softmax`: Converts logits into probabilities.

2. **Data Flow**:
   - Input data is passed through each layer in sequence.
   - For example, in the above models:
     - Input $$x$$ passes through `fc1`, then `relu1`, then `fc2`, and finally `softmax`.

3. **Output**:
   - The final output is determined by the last layer (e.g., probabilities after softmax).

---

### **Example with Training**

Here's how you can train a model defined using `nn.Sequential`:

```python
# Example dataset and dataloader (e.g., MNIST)
from torch.utils.data import DataLoader, TensorDataset

# Dummy data: 100 samples, each with 784 features (e.g., flattened images)
X = torch.randn(100, 784)

dataloader = DataLoader(dataset, batch_size=10)

optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# Training loop
for epoch in range(5):  # Train for 5 epochs
    for batch_X, batch_y in dataloader:
        optimizer.zero_grad()       # Zero gradients
        outputs = model(batch_X)    # Forward pass
        loss = criterion(outputs, batch_y)  # Compute loss
        loss.backward()             # Backward pass
        optimizer.step()            # Update weights

    print(f"Epoch {epoch+1}, Loss: {loss.item()}")
```

### **When to Use `nn.Sequential` vs Subclassing `nn.Module`**

| Feature                     | **`nn.Sequential`**                              | **Subclassing `nn.Module`**                  |
|-----------------------------|--------------------------------------------------|----------------------------------------------|
| Simplicity                  | Best for linear architectures                    | Suitable for complex architectures           |
| Flexibility                 | Limited (no branching or shared layers)          | Highly flexible                              |
| Customization               | Cannot customize data flow                       | Fully customizable                           |
| Use Case                    | Simple feed-forward networks                     | Networks with branching or multiple inputs/outputs |

For example:
- Use `nn.Sequential` for simple models like multi-layer perceptrons (MLPs).
- Use subclassing for more advanced models like CNNs or RNNs with non-linear data flow.

## **Using Pretrained Models for Transfer Learning**

Transfer learning involves modifying a pretrained model to adapt it to a new task. For example, replacing the final classification layer of ResNet-50 to classify a custom dataset with 10 classes:

```python
import torch
from torchvision import models
import torch.nn as nn

# Load the pretrained ResNet-50 model
model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)

# Modify the final fully connected layer for 10 classes (instead of 1000)
num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 10)  # Replace the fully connected layer

# Set the model to training mode
model.train()

# Example input (batch of images with shape [batch_size, channels, height, width])
example_input = torch.randn(8, 3, 224, 224)  # Batch size of 8, RGB images of size 224x224

# Forward pass
output = model(example_input)
print(output.shape)  # Output shape: [8, 10]
```

---

## **Steps for Transfer Learning**
1. **Load Pretrained Model**: Use `torchvision.models` to load a model with pretrained weights.
2. **Modify Output Layer**: Replace the final classification layer with one matching the number of classes in your dataset.
3. **Freeze Layers (Optional)**: Freeze earlier layers if you want to retain their weights and only train the new layers.
   ```python
   for param in model.parameters():
       param.requires_grad = False  # Freeze all layers

   # Unfreeze only the final layer
   model.fc.requires_grad = True
   ```
4. **Train on New Dataset**: Use your custom dataset and train the modified model.

| **Training Loop**              | Training loop must be implemented manually.      | Built-in `model.fit()` simplifies training.     |

Both frameworks are powerful; PyTorch offers more flexibility and control, while TensorFlow/Keras is easier for rapid prototyping.
